```python
import pandas as pd
from sklearn.svm import SVC
from sklearn.feature_selection import SelectFromModel
# Convert 'Access Time' column to datetime
data['Access Time'] = pd.to_datetime(data['Access Time'])

# Extract year, month, day, hour, and minute as numerical features
data['Year'] = data['Access Time'].dt.year
data['Month'] = data['Access Time'].dt.month
data['Day'] = data['Access Time'].dt.day
data['Hour'] = data['Access Time'].dt.hour
data['Minute'] = data['Access Time'].dt.minute



# Define feature columns and target variable
X = data.drop('label', axis=1)  # Features
y = data['label']  # Target variable


# Use a linear SVM as the base estimator for feature selection
svm_clf = SVC(kernel='linear')
svm_clf.fit(X, y)

# Use SelectFromModel to perform feature selection based on SVM
selector = SelectFromModel(estimator=svm_clf)
selector.fit(X, y)

# Get selected features
selected_features = X.columns[selector.get_support()]

print("Selected Features:")
print(selected_features)
```
